```python
df = yf.Ticker('HDFCBANK.NS').history(period='5y').reset_index()
df.columns = df.columns.str.lower()
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
df = df[['date','close','volume']]
df['close']=df['close'].round(2)
display(
  df
)
```

```python
dropdown = df[df['date'].str.contains('2022')]
```

```python
df['close'].cummax().plot()
```

```python
df['close'].pct_change().plot.bar()
```

```python
(( df['close'] -df['close'].cummax() ) / df['close'].cummax()).plot.bar()
```
